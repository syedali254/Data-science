```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns


df=pd.read_csv("Titanic-Dataset.csv")

df.head()
print(np.mean(df["Age"]))

mn=np.mean(df["Age"])


sns.histplot(x="Age",data=df,bins=[i for i in range(0,80,5)])
plt.plot([mn for i in range(300)],[i for i in range(300)])
plt.show()
```

```python
sec_a=np.array([75,65,73,68,72,76])
sec_b=np.array([90,47,43,96,93,51])
no=np.array([1,2,3,4,5,6])
mean_a=np.mean(sec_a)
mead_b=np.mean(sec_b)
```

```python
plt.figure(figsize=(10,3))
plt.scatter(sec_a,no,label="Sec A")
plt.scatter(sec_b,no,label="Sec B")
plt.legend()
plt.show()
```

```python
mad_a=np.sum(np.abs(sec_a-mean_a))/len(sec_a)
print(mad_a)
```

```python
mad_b=np.sum(np.abs(sec_b-mead_b))/len(sec_b)
print(mad_b)
```
