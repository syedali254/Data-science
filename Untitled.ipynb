```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df2=pd.read_csv("tip.csv")
print(df2.head(5))
```

```python
print(df2.isnull().sum())
```

```python
df2.info()
```

```python
df2.select_dtypes(["float64","int64"]).corr()
```

```python
df2.select_dtypes(["float64","int64"]).cov()
```

```python
import numpy as np
np.cov(df2["total_bill"],df2["tip"])  #covariance b/w 2 columns using np
```

```python
cor_data=np.corrcoef(df2["total_bill"],df2["tip"])
```

```python
print(cor_data)
plt.figure(figsize=(4,3))
sns.heatmap(cor_data,annot=True)
plt.show()
```

```python
plt.figure(figsize=(4,3))
cov_data=np.cov(df2["total_bill"],df2["tip"])
sns.heatmap(cov_data,annot=True)
plt.show()
```
